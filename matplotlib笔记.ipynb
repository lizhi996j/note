## Matplotlib

### 关于图中的坐标
有三种坐标：
- 数据坐标：每个数据点在图中的位置，由x和y坐标决定。
- 轴坐标：每个轴在图中的位置，由轴的起点和终点决定。
- 图坐标：整个图在图中的位置，由图的起点和终点决定。

| 坐标类型 | 函数/方法 | 坐标范围 | 典型用途 |
|----------|-----------|----------|----------|
| 数据坐标 | plt.text / ax.text | 按数据值 | 在数据点处标注 |
| 轴坐标 | ax.text(..., transform=ax.transAxes) | [0,1] | 在子图固定位置放置文字、图例等 |
| 画布坐标 | fig.text | [0,1] | 全局标题、水印、注释框 |


```python
import matplotlib.pyplot as plt

x = [0, 1, 2]
y = [1, 4, 9]
plt.plot(x, y)

# 在数据坐标 (1, 4) 处放文字
plt.text(1, 4, "这里是 (1,4)", fontsize=12)

plt.show()
```

### 绘制线形图
matplotlib.pyplot.plot(x, y, format_string, **kwargs)函数：  
- format_string：线条的样式、颜色和标记.   
    - 颜色字符：'b' 蓝色，'m' 洋红色，'g' 绿色，'y' 黄色，'r' 红色，'k' 黑色，'w' 白色，'c' 青绿色，'#008000' RGB 颜色符串。多条曲线不指定颜色时，会自动选择不同颜色。  
    - 线型参数：'‐' 实线，'‐‐' 破折线，'‐.' 点划线，':' 虚线。  
    - 标记字符：'.' 点标记，',' 像素标记(极小点)，'o' 实心圈标记，'v' 倒三角标记，'^' 上三角标记，'>' 右三角标记，'<' 左三角标记...等等
- 用于设置更多的绘图参数，比如标签、标题、线宽、颜色等


```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 100])

plt.plot(xpoints, ypoints)
plt.show()
plt.plot(xpoints,ypoints,'y:o')
plt.show()
```

#### 定义标记的内部颜色，边框颜色和大小
- markersize，简写为 ms：定义标记的大小。
- markerfacecolor，简写为 mfc：定义标记内部的颜色。
- markeredgecolor，简写为 mec：定义标记边框的颜色。


```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([6,2,13,10])

plt.plot(y,marker="o",ms=20,mfc="b",mec="r")
```

自定义内部和边框颜色  
颜色代码（Hex Color Code）是一种用六位十六进制数字来表示 RGB 颜色的方法，格式为 #RRGGBB，其中：

RR：表示红色通道（Red）的强度，范围从 00（最小，十进制 0）到 FF（最大，十进制 255）；

GG：表示绿色通道（Green）的强度，同样从 00 到 FF；

BB：表示蓝色通道（Blue）的强度，同样从 00 到 FF。

举例来说，#4CAF50 可以拆解为：

红色 4C → 十六进制 4C = 十进制 76

绿色 AF → 十六进制 AF = 十进制 175

蓝色 50 → 十六进制 50 = 十进制 80

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4C0F50')
```

##### 如果x的范围不同会取并集  
当你用 plt.xlim(a, b) 手动限制横轴范围为 [a, b] 时，任何落在此区间之外的数据点都不会被绘制到图上——它们被“剪裁”掉了。

```python
import matplotlib.pyplot as plt

# 第一条线的 x1 范围是 [0,1,2,3]
x1 = [0, 1, 2, 3]
y1 = [1, 4, 9, 16]

# 第二条线的 x2 范围是 [1,3,5]
x2 = [1, 3, 5]
y2 = [2, 6, 3]

plt.plot(x1, y1, label='y1')
plt.plot(x2, y2, label='y2')
plt.legend()
plt.show()
```

##### 绘制x轴不同但共享y轴的图 
生成一个ax2, ax2与ax1共享y轴但独立x轴

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twiny()   # 生成一个共享 y 轴但独立 x 轴的坐标系

ax1.plot(x1, y1, color='blue', label='y1 on ax1')
ax2.plot(x2, y2, color='red',  label='y2 on ax2')

# 分别设置两个坐标系的 x 轴标签
ax1.set_xlabel('x1 in [0,3]')
ax2.set_xlabel('x2 in [1,5]')

# 可以单独设置它们的 xlim
ax1.set_xlim(0, 3)
ax2.set_xlim(1, 5)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()
```

### 显示轴和标题
- 轴和标题的位置：
    - title() 方法提供了 loc 参数来设置标题显示的位置，可以设置为: 'left', 'right', 和 'center'， 默认值为 'center'。

    - xlabel() 方法提供了 loc 参数来设置 x 轴显示的位置，可以设置为: 'left', 'right', 和 'center'， 默认值为 'center'。

    - ylabel() 方法提供了 loc 参数来设置 y 轴显示的位置，可以设置为: 'bottom', 'top', 和 'center'， 默认值为 'center'。

#### matplotlib.pyplot.title(label, fontdict=None, loc='center', pad=None, **kwargs)  
- label: 标题文本
- fontdict:字体字典，可以设置：
    - fontsize（字号，如 12、'large' 等）
    - fontweight（粗细，如 'bold'、'semibold'）
    - fontstyle（样式，如 'italic'）
    - fontfamily（字体族，如 'serif'、'sans-serif'）
    - color（颜色，如 'blue'、'#333333'）
可以如下面例子一样传递字典形式，也可以直接传递参数
- loc: 标题位置，可选值：'center'、'left'、'right'
- pad: 标题与图的距离，默认值为 None
- **kwargs: 其他参数，如 backgroundcolor、bbox、rotation等 <font color='red'>关于backgroundcolor和bbox中facecolor的区别没太懂</font>

```python
import matplotlib.pyplot as plt

# 准备数据
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y)

# 使用 fontdict 和 loc、pad、额外 kwargs
ax.set_title(
    "Example",
    fontdict={
        'fontsize': 16,
        'fontweight': 'bold',
        'fontfamily': 'sans-serif',
        'color': '#2E86AB'
    },
    loc='left',
    pad=20,
    backgroundcolor='#000000',
    bbox={'facecolor': '#FFFFFF', 'edgecolor': '#2E86AB', 'boxstyle': 'round,pad=0.3'},
    rotation=0
)

plt.show()
```

### 设置刻度线
两种组合：
1. 手动设置刻度位置和标签
- Axes.set_xticks(self, ticks, *, minor=False)
- Axes.set_yticklabels(
    labels, 
    *, 
    fontdict=None, 
    minor=False, 
    **kwargs
)
2. 结合locator和formatter自动设置刻度位置和标签
- Axes.xaxis.set_major_locator(locator)
    locator的种类：
    - AutoLocator: 自动选择最佳主刻度
    - MaxNLocator(n): 在坐标轴范围内最多生成 n 条主刻度，刻度间距尽量“整洁”（如整数、四舍五入后的小数）
    - MultipleLocator(base): 在坐标轴范围内，每隔 base 单位生成一个刻度
    - IndexLocator(base, offset): 用于IndexLocator不关心数据的实际数值大小，只按“第几个点”计算位置的情况，从offset开始，每隔base个单位生成一个刻度
- Axes.xaxis.set_major_formatter(formatter)
    formatter的种类：
    - FormatStrFormatter(fmt): 	按 Python 格式化字符串（如 '%0.2f'、'%d%%'）来渲染标签
    - PercentFormatter(xmax=1.0)	将刻度值按百分比显示（0–1 映射为 0%–100%）
    - FuncFormatter(func): 使用自定义函数来渲染标签
    - FixedFormatter(values): 使用固定值列表来渲染标签
    - DateFormatter(fmt): 使用日期格式化字符串来渲染标签
3. ax.tick_params(axis='both', which='major', **kwargs) 是从“样式/显示属性”层面，统一控制刻度线（ticks）和刻度标签（tick labels）的外观（线长、线宽、方向、颜色、字体大小等）。  
4. 获得tick的location和label  
    tick_locations = ax.get_xticks()  
    tick_labels = ax.get_xticklabels()  # 返回的是text对象  
    print(tick_locations)  
    print(tick_labels)  
    

#### 第一种方式的例子：

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [1, 3, 2, 5, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y, marker='o')

# 将 x 轴刻度强制设为 [0, 1, 2, 3, 4, 5]
ax.set_xticks([0, 1, 2, 3, 4, 5])

# 将 y 轴添加次刻度 [1.5, 2.5, 3.5, 4.5]
ax.set_yticks([1.5, 2.5, 3.5, 4.5], minor=True)

# 可以配合格式化函数或标签显示
ax.set_yticklabels(['A','B','C','D'], minor=True)

plt.show()
```

#### 第二种方式的例子

##### 例1：maxlocatror 和 multiplelocator 的结合使用

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoLocator, MaxNLocator, MultipleLocator, FixedLocator

fig, ax = plt.subplots()

x = range(11)
y = [i**1.5 for i in x]
ax.plot(x, y, marker='o')

# 1. 自动选取“最佳”主刻度
ax.xaxis.set_major_locator(MultipleLocator(2))

# 2. Y 轴最多 5 条主刻度，数字更整洁
ax.yaxis.set_major_locator(MaxNLocator(5))


# 调整刻度样式
ax.tick_params(which='major', length=8, width=1.2)

plt.show()
```

##### 例2：indexlocator和fixedformatter的例子(用于处理分类数据)

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import IndexLocator, FixedFormatter

# 分类与对应数值
categories = ['A','B','C','D','E']
values     = [10, 20, 15, 25, 18]
x_indices  = list(range(len(categories)))  # 底层位置：0,1,2,3,4

fig, ax = plt.subplots()

# 画折线
ax.plot(x_indices, values, marker='o')

# —— 1) 用 IndexLocator 自动每隔 1 个索引打主刻度 —— #
ax.xaxis.set_major_locator(IndexLocator(base=1, offset=0))

# —— 2) 用 FixedFormatter 把这些刻度映射成分类名 —— #
ax.xaxis.set_major_formatter(FixedFormatter(categories))

# 样式微调
ax.tick_params(which='major', length=6)
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('IndexLocator + FixedFormatter 示例')

plt.tight_layout()
plt.show()
```

##### 例3：一个综合使用locator和formatter的例子

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MultipleLocator
import numpy as np

# 生成示例数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 创建图表
fig, ax = plt.subplots()

ax.plot(x, y)

# 设置 x 轴主刻度为每 2 单位一个
ax.xaxis.set_major_locator(MultipleLocator(2))
# 设置 x 轴次刻度为每 1 单位一个
ax.xaxis.set_minor_locator(MultipleLocator(1))

# 自定义 x 轴标签格式：保留一位小数
ax.xaxis.set_major_formatter(FuncFormatter(lambda val, pos: f"{val:.1f}"))

# 自定义 y 轴标签格式：以百分比显示
ax.yaxis.set_major_formatter(FuncFormatter(lambda val, pos: f"{val*100:.0f}%"))

# 添加标题与网格
ax.set_title("示例: Locator 和 Formatter")
ax.grid(True)

plt.show()
```

##### matplotlib.ticker.LogLocator(base=10.0, subs=None, numticks=15, *, minor=False)与.logformatter
- LogLocator参数：  

| 参数 | 类型 | 默认值 | 说明 |
|------|------|--------|------|
| base | float | 10.0 | 对数的底，例如 10 表示以 10 为底的对数刻度（常用）。 |
| subs | sequence/None | None | 子刻度的位置（子倍数）就是不同量级之间主刻度的数量，例如 [1.0, 2.0, 5.0] 表示显示 log base 上的这些倍数。若为 None，主刻度将自动生成。 |
| numticks | int | 15 | 最大刻度数量（是个建议值，不一定严格遵守）。 |
| minor | bool | False | 是否是用于次刻度（minor ticks）。若为 True，表示这是一个次刻度定位器。 |  

- LogFormatter参数：  

| 参数 | 类型 | 默认值 | 说明 |
|------|------|--------|------|
| base | float | 10.0 | 对数的底，例如 10 表示以 10 为底的对数刻度（常用）。 |
| labelOnlyBase | bool | False | 是否只在底数的整数次幂处显示标签（如 10^0, 10^1, 10^2）。若为 True，则只显示这些位置的标签。 |
| minor_thresholds | tuple | (1, 0.4) | 控制次刻度标签显示的阈值。第一个值控制主刻度间隔，第二个值控制次刻度密度。 |
| linthresh | float | None | 线性阈值，用于对称对数刻度。 |  
- 如果要用科学计数法显示label，则需要使用LogFormatterSciNotation


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import LogLocator,LogFormatter

# 数据
x = np.logspace(0.1, 3, 400)  # x 从 10^0.1 到 10^2，400 个点
y = np.sqrt(x)

# 创建图形
fig, ax = plt.subplots()
ax.plot(x, y)

# 设置对数坐标轴
ax.set_xscale('log')
ax.set_yscale('log')

# 使用 LogLocator 设置主刻度和次刻度
ax.yaxis.set_major_locator(LogLocator(base=10.0, numticks=10,subs=[1.0,2.0,5.0]))
ax.yaxis.set_major_formatter(LogFormatter(labelOnlyBase=False,minor_thresholds=(np.inf,np.inf)))

# 添加网格以更好地观察刻度
# ax.grid(True, which='both', linestyle='--', linewidth=0.5)

ax.set_title("LogLocator 示例")
plt.show()
```

#### ax.tick_params(axis='both', which='major', **kwargs)的用法：

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, label='sin(x)')

# 应用 tick_params
# 对主刻度进行设置：字体大小、字体颜色、刻度线长度、线宽、刻度线方向和颜色
ax.tick_params(
    axis='both',       # 设置 x 和 y 轴
    which='major',     # 主要刻度
    labelsize=12,      # 刻度标签字体大小
    labelcolor='blue', # 刻度标签颜色
    length=8,          # 刻度线长度
    width=2,           # 刻度线宽度
    direction='in',    # 刻度线朝内绘制
    colors='red'       # 刻度线颜色
)

# 对次刻度进行设置：开启次刻度并设置长度和颜色
ax.minorticks_on()
ax.tick_params(
    axis='both',
    which='minor',
    length=4,
    colors='green'
)

ax.set_title("示例：ax.tick_params 的使用")
ax.legend(loc='upper left')

plt.show()

```
